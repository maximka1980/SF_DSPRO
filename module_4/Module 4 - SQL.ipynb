# 5.3. SELECT FROM SELECT


```python
Задание 5.3.1
/*
Напишите запрос, который найдет водителя, совершившего наибольшее количество доставок одному клиенту. 
В выводе должна быть одна строка, которая содержит имя водителя (first_name), имя клиента и количество доставок водителя этому клиенту. 
*/
select
    a.first_name,
    a.cust_name,
    count(a.ship_id) ships_total
from
    (
    select
        d.first_name,
        cs.cust_name,
        s.ship_id
    from
        shipping.shipment s
            join shipping.driver d on s.driver_id = d.driver_id
            join shipping.customer cs on s.cust_id = cs.cust_id
    ) a
group by
    a.first_name,
    a.cust_name
order by
    ships_total desc
limit 1
```

```python
Задание 5.4.1 

with PL as
(
select *
from
shipping.shipment
where ship_date < to_date('01.02.2017', 'DD.MM.YYYY')
),
NL as
(
select *
from
shipping.shipment
where ship_date >= to_date('01.02.2017', 'DD.MM.YYYY')
)
select 'Paolo Lorenzo' as Director, count(pl.ship_id) ship_count, count (distinct pl.driver_id) driver_count, count (distinct pl.cust_id) cust_count, sum(pl.weight) sum_weight
from PL
union all
select 'Nicco Lorenzo', count(nl.ship_id), count (distinct nl.driver_id), count (distinct nl.cust_id), sum(nl.weight)
from NL
order by 1
```
